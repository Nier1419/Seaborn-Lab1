## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(data=df, x="sex", y="survived");
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(data=df, x="survived", col="embarked", kind="count", height=5, aspect=1);
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
#adjusting the size of boxplot (changes style too)
sns.set(rc={"figure.figsize":(15, 8)})
#to return the style as it was before
sns.set_style("white")

sns.boxplot(data=df, x="sex", y="age", hue="pclass");
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
#we need to pass the dataframe we want to draw correlaization from to "corr()" function 
correlizedData=df[['sibsp', 'parch', 'age', 'fare', 'survived']].corr()
#heatmap(dataToGetCorrelationFrom, putVluesInsideSquares, Show2NumbersAfterThePoint)
sns.heatmap(correlizedData, annot=True, fmt=".2f");
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.factorplot(x ='survived', y ='sibsp', data = df);
```
